```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","What you Know about India ?"),
        ("user","{input}")
    ]
)
prompt
```

```python
chain = prompt|llm

response = chain.invoke({"input":"Can you tell me about India ?"})
print(response)
```
